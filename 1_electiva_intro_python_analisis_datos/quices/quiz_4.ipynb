### **Pregunta 4 – Visualización con Matplotlib (1.0 punto)**

Observe el siguiente código:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Relación entre x e y")
plt.show()
```

a) ¿Qué tipo de relación se observa entre `x` e `y`?
b) ¿Qué información comunica este gráfico?
