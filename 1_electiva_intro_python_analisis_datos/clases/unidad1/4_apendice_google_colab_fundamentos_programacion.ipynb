## 1. ¿Qué es Google Colab?

Google Colab (Colaboratory) es un entorno gratuito basado en la nube que permite ejecutar código Python directamente desde el navegador.

### Ventajas
- No requiere instalación local.
- Integración con Google Drive.
- Uso de CPU y GPU.
- Compartición en tiempo real.
